To import the necessary packages:

```python
#Ignore warnings
import warnings
warnings.filterwarnings("ignore")

import xarray as xr
import cartopy.crs as ccrs
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import matplotlib.pyplot as plt
import numpy as np
```

Xarray provides a convenient interface import data from NetCDF files. To load data file:

```python
#Open the dataset and print out metadeta
ncfile = '~/project/mc2019/public/T2-dataset/surface/air.2m.gauss.2018.nc'
ds = xr.open_dataset(ncfile)
# Open surface wind data
ncfile2 = '~/project/mc2019/public/T2-dataset/multiple_levels/uwnd.2018.nc'
ds2 = xr.open_dataset(ncfile2)
ncfile3 = '~/project/mc2019/public/T2-dataset/multiple_levels/vwnd.2018.nc'
ds3 = xr.open_dataset(ncfile3)
```

To retrive the fields of "lat" and "lon". These two variables have defined the grid of this NetCDF file.

lat/lon is the center value of the pixel

Retrieve fields for uwnd and vwnd, and select 1000 hPa pressure wind speed

```python
lat = ds['lat']
lon = ds['lon']
lat1 = ds2['lat']
lon1 = ds2['lon']

air = ds['air']
uwnd = ds2['uwnd'].sel(level=1000)
vwnd = ds3['vwnd'].sel(level=1000)
```

# Question 3

## (a) Plot the surface temperature anomaly for January 2018 relative to the long-term mean for January (1981-2010). 

```python
ds5 = xr.open_dataset('public/T2-dataset/long_term_144x73/air.mon.ltm.nc')
ds6 = xr.open_dataset('public/T2-dataset/long_term_144x73/air.mon.mean.nc')
ds5,ds6
```

```python
air_Jan_ltm = ds5['air'].sel(level=1000).where(ds5.time.dt.month==1,drop=True).squeeze(dim='time')
air_Jan_mean = ds6['air'].sel(level=1000).where(ds6.time.dt.month==1,drop=True).mean(dim='time')
air_Jan_ann = air_Jan_mean-air_Jan_ltm

air_Jul_ltm = ds5['air'].sel(level=1000).where(ds5.time.dt.month==6,drop=True).squeeze(dim='time')
air_Jul_mean = ds6['air'].sel(level=1000).where(ds6.time.dt.month==6,drop=True).mean(dim='time')
air_Jul_ann = air_Jul_mean-air_Jul_ltm
```

```python
fig,ax = plotmap()
pltdata=air_Jan_ann.values
levelx = np.arange(-3,3,0.5)
plotx_plot = ax.contourf(lon1,lat1,pltdata,levels=levelx,cmap='bwr',zorder=2,transform=ccrs.PlateCarree())

#Add colorbar
cbar = plt.colorbar(plotx_plot,shrink=0.6,pad=0.035)
cbar.ax.tick_params(labelsize=13)

ax.set_title('Air Temperature Annomaly (degC) 2018 January',fontsize=15)
```

## (b) Repeat the process for July 2018.

## Calculate the zonal mean surface air temperature for January and July 2018. Plot both data in one plot. This would be a 1D plot, with latitude as the x-axis and temperature as the y-axis. 

```python
air_zonal_Jan = ds6['air'].sel(level=1000,time=slice('2018-01-01','2018-12-31')).where(ds6.time.dt.month==1).mean(dim='time').mean(dim='lon')
air_zonal_Jul = ds6['air'].sel(level=1000,time=slice('2018-01-01','2018-12-31')).where(ds6.time.dt.month==7).mean(dim='time').mean(dim='lon')
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.plot(ds6.lat,air_zonal_Jan,color='darkcyan',lw=3)
ax.plot(ds6.lat,air_zonal_Jul,color='salmon',lw=3)
ax.set_xlim(-90,90)
ax.set_xticks(np.arange(-90,91,30))
ax.set_xticklabels(['90°S','60°S','30°S','EQ','30°N','60°N','90°N'],
                  fontsize=15)
ax.set_yticklabels(ax.get_yticks(),fontsize=15)
ax.set_ylabel('Latitude',fontsize=15)
ax.set_xlabel('Temperature ℃',fontsize=15)
ax.legend(['Januaray','July'], fontsize=15)
ax.set_title('2018 Zonal Mean Air Temperature', fontsize=18)
ax.axvline(0,color='k')
ax.grid()
```
